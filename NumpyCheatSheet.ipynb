```python
# prompt: create a cheat sheet for numpy i, atleast 50 functions

import numpy as np

# Array Creation
np.array([1, 2, 3])  # Create an array from a list
np.zeros(5)  # Create an array of zeros
np.ones(3)  # Create an array of ones
np.arange(0, 10, 2)  # Create an array with a range of values
np.linspace(0, 1, 5) # Create an array with evenly spaced numbers
np.empty(4)  # Create an uninitialized array
np.full(3, 7) # Create an array filled with a specific value
np.eye(3)  # Create an identity matrix
np.random.rand(2, 3)  # Generate random numbers from a uniform distribution
np.random.randn(2, 3)  # Generate random numbers from a standard normal distribution
np.random.randint(0, 10, size=5) # Generate random integers within a range
np.random.choice([1,2,3], size=4) # Randomly sample from a given array
np.random.seed(42) # Set random seed for reproducibility


# Array Manipulation
np.reshape(np.array([1,2,3,4]), (2,2)) # Reshape an array
np.transpose(np.array([[1,2],[3,4]])) # Transpose an array
np.concatenate((np.array([1,2]), np.array([3,4]))) # Concatenate arrays
np.vstack((np.array([1,2]), np.array([3,4]))) # Vertical stack
np.hstack((np.array([1,2]), np.array([3,4]))) # Horizontal stack
np.split(np.array([1,2,3,4]), 2)  # Split an array into sub-arrays
np.delete(np.array([1,2,3]), 1) # Delete element at a specific index
np.insert(np.array([1,2,3]), 1, 4) # Insert element at a specific index
np.append(np.array([1,2,3]), 4) # Append element at the end


# Array Attributes
np.array([1, 2, 3]).ndim # Number of dimensions
np.array([1, 2, 3]).shape # Dimensions of the array
np.array([1, 2, 3]).size  # Total number of elements
np.array([1, 2, 3]).dtype  # Data type of elements


# Mathematical Operations
np.add(np.array([1,2]), np.array([3,4])) # Element-wise addition
np.subtract(np.array([1,2]), np.array([3,4])) # Element-wise subtraction
np.multiply(np.array([1,2]), np.array([3,4])) # Element-wise multiplication
np.divide(np.array([1,2]), np.array([3,4])) # Element-wise division
np.dot(np.array([1,2]), np.array([3,4])) # Dot product
np.sum(np.array([1,2,3])) # Sum of elements
np.mean(np.array([1,2,3])) # Mean of elements
np.std(np.array([1,2,3])) # Standard deviation of elements
np.min(np.array([1,2,3])) # Minimum element
np.max(np.array([1,2,3])) # Maximum element
np.argmin(np.array([1,2,3])) # Index of minimum element
np.argmax(np.array([1,2,3])) # Index of maximum element

# Linear Algebra
np.linalg.inv(np.array([[1,2],[3,4]])) # Inverse of a matrix
np.linalg.det(np.array([[1,2],[3,4]]))  # Determinant of a matrix
np.linalg.eig(np.array([[1,2],[3,4]]))  # Eigenvalues and eigenvectors


#Other Useful Functions
np.unique(np.array([1,1,2,2,3])) # Find unique elements
np.where(np.array([1,2,3]) > 1) # Find indices where condition is true
np.sort(np.array([3,1,2])) # Sort the array
np.argsort(np.array([3,1,2])) # Indices that would sort the array
np.any(np.array([1,0,0]) > 0) # Checks if any element satisfies condition
np.all(np.array([1,2,3]) > 0) # Checks if all element satisfies condition
```

```python
# prompt: give more functions of nummpy for data preprocessing

import numpy as np

# ... (Your existing code)

# Outlier detection using IQR
data = np.array([2, 3, 4, 6, 8, 9, 10, 45])
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data < lower_bound) | (data > upper_bound)]
print("Outliers using IQR:", outliers)


#Normalization (Min-Max scaling)
data = np.array([1, 5, 10, 15, 20])
normalized_data = (data - np.min(data)) / (np.max(data) - np.min(data))
print("Normalized data:", normalized_data)


#Standardization (Z-score normalization)
data = np.array([1, 5, 10, 15, 20])
standardized_data = (data - np.mean(data)) / np.std(data)
print("Standardized data:", standardized_data)


#Binarization
data = np.array([1, 5, 10, 15, 20])
threshold = 10
binarized_data = np.where(data > threshold, 1, 0)
print("Binarized data:", binarized_data)


#Imputation (using mean)
data = np.array([1, 5, np.nan, 15, 20])
mean_value = np.nanmean(data)
imputed_data = np.nan_to_num(data, nan=mean_value)
print("Imputed data:", imputed_data)

# One-hot encoding (example for a single feature)
labels = np.array(['red', 'green', 'blue', 'red'])
unique_labels = np.unique(labels)
one_hot_encoded = np.zeros((len(labels), len(unique_labels)))
for i, label in enumerate(labels):
  one_hot_encoded[i, np.where(unique_labels == label)] = 1
print("One-hot encoded labels:\n",one_hot_encoded)
```
